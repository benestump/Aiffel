하지만 저의 경우에는 jupyter notebook을 중단하고 다시 시작하는 경우에는 default font로 변경되는 문제가 있었습니다   
그래서 /home/aiffel/.ipython/profile_default/startup/ 아래에 00-first.py 파일을 만들어주었습니다.    
00-first.py 파일은 jupyter notebook이 실행되기 전에 해당파일이 가장 먼저 실행되어 설정을 할 수 있습니다.   
해당 파일 안에는 아래의 코드를 넣어두었습니다. 

```python
import matplotlib as mpl
import matplotlib.font_manager as fm

# 경로는 위에서 확인한 경로명을 입력하여야 합니다. 
font_path = '/home/benestump/anaconda3/envs/aiffel/lib/python3.7/site-packages/matplotlib/mpl-data/fonts/ttf/D2Coding-Ver1.3.2-20180524.ttf'
font_name = fm.FontProperties(fname=font_path).get_name()

mpl.rc('font', family=font_name)
```

이후에는 jupyter notebook을 종료하고 다시 시작하더라도 D2Coding font로 설정되어 있습니다. 또한 해당 파일에 자주 사용하는 코드를 미리 입력하여 두는 방식으로 활용이 가능합니다.
